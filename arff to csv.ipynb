```python
import pandas as pd
from scipy.io import arff
data = arff.loadarff('D:\ARFF to CSV/breast.w.arff')
train= pd.DataFrame(data[0])
train.head()
```

```python
train.info()
```

```python
catCols = [col for col in train.columns if train[col].dtype=="O"]
catCols[:5]
```

```python
train[catCols]=train[catCols].apply(lambda x: x.str.decode('utf8'))
train.head()
```

```python
train.to_csv('./breast.w.csv',index=False)
```
